- **Assignment : Basics of Data Analytics

**

- 1. Define the role of a Data Analyst. What value do they bring to an organization?

  - A Data Analyst is a professional who collects, cleans, analyzes, and interprets data to help an organization make better decisions.
  - They turn raw numbers into meaningful insights.

- Value they bring:

Help companies understand trends and patterns

Support business decisions with facts, not guesswork

Improve efficiency by identifying problems and opportunities

Create dashboards and reports for management

Predict results using data-driven insights

2. List three tools commonly used by Data Analysts and what each is used for.
1. Excel

Used for data cleaning, calculations, PivotTables, charts, and quick analysis.

2. SQL

Used to extract data from databases, filter information, join tables, and answer business queries.

3. Power BI / Tableau

Used to create interactive dashboards, visualizations, and business reports.

3. Write the end-to-end analytics workflow in the correct order + explain each step.
1. Data Collection

Gather data from sources like databases, surveys, files, websites, etc.

2. Data Cleaning

Fix missing values, remove duplicates, correct errors, and standardize the data.

3. Data Exploration

Understand the data using summary statistics, graphs, and trend analysis.

4. Data Analysis / Modeling

Apply formulas, SQL queries, or statistical/ML models to extract insights.

5. Visualization & Reporting

Create dashboards and charts to communicate insights clearly.

6. Decision Making

Business teams use insights to take actions and improve performance.

4. What is Prompt Engineering? Explain any two types of prompts used in Generative AI.
Prompt Engineering:

It is the technique of designing clear and effective instructions for AI models to get accurate results.
Better prompts = better AI output.

Types of Prompts:
1. Instruction Prompt

Directly tells the AI what to do.
Example: “Explain machine learning in simple words.”

2. Context Prompt

Provides background information so AI can give a better response.
Example: “Here is sales data for 3 months. Summarize the trend.”

5. Key differences between a Business Analyst and a Data Analyst.
Business Analyst	Data Analyst
Focuses on business needs and processes	Focuses on data and numbers
Understands requirements from stakeholders	Cleans, analyzes, and visualizes data
Creates business strategies and solutions	Creates reports, dashboards, and insights
Works more with workflows & business planning	Works more with SQL, Excel, BI tools
Goal: Improve business operations	Goal: Provide data-driven decisions
6. Explain any three AI-powered ETL tools and how they are used.

Uses AI to detect errors, improve data quality, and optimize workflows.

2. Informatica Intelligent Cloud Services (IICS)

1. Healthcare

2. Retail / E-commerce

Customer recommendations, predicting sales, tracking buying behavior.

3. Banking & Finance

8. Explain the evolution of analytics from Descriptive to Generative AI.
1. Descriptive Analytics (Past)

Tells what happened.
Example: Monthly sales reports.

2. Diagnostic Analytics

Explains why it happened.
Example: Finding reasons for low sales.

3. Predictive Analytics

Predicts what will happen next.
Example: Sales forecasting.

4. Prescriptive Analytics

Suggests what action to take.
Example: Best pricing strategy or product mix.

5. Generative AI (Today)

AI creates new insights, images, reports, strategies, and content based on patterns learned from data.
It automates analysis, gives recommendations, and enhances productivity.
